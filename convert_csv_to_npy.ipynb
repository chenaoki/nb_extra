```python
import glob, os
import numpy as np
```

```python
dirs = glob.glob(src_dir+'201*/e*')

dirs
```

```python
for d in dirs:
    print(d)
    files = sorted(glob.glob(os.path.join(d, "phase/*.csv")))
    data = np.zeros((len(files), 96, 90))
    for i, f in enumerate(files):
        im = np.genfromtxt(f, delimiter=',', filling_values=0).astype(np.float64)
        im = np.deg2rad(im)
        H,W = im.shape
        data[i,:H,:W] = im
    data[np.isnan(data)] = 0
    np.save(d, data)
```
